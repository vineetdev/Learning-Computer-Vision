```python
qr = qrcode.QRCode(version=1,
                   error_correction=qrcode.constants.ERROR_CORRECT_1,
                   box_size = 10,
                   border = 5)

data_to_encode = "https://www.linkedin.com/in/vineetkrsrivastava"
qr.add_data(data_to_encode)
qr.make(fit=True)

img = qr.make_image(fill_color="black", back_color="white")
img.save("trialQR.png")
```
